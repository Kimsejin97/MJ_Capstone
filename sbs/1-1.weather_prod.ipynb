```python
import pandas as pd
a = pd.read_csv('./채소생산량_근채류__20200506160919.csv', encoding='euc-kr')
b = pd.read_csv('./채소생산량_엽채류__20200506150012.csv', encoding='euc-kr')
c = pd.read_csv('./채소생산량_조미채소__20200506145121.csv', encoding='euc-kr')
```

```python
df = pd.concat([a,b,c],axis=1).T
df = df.rename(columns=df.iloc[0])
# df.drop(df.index[0], inplace=True)
df = df[df['시도별'] != '10a당 생산량 (kg)']
df['시도별'] = df['시도별'].str.replace(pat=':면적 (ha)', repl='', regex=False)
df['시도별'] = df['시도별'].str.replace(pat='생산량 (톤)', repl='Ton', regex=False)
df = df.reset_index(drop=True)
s = sorted(set(df['시도별']))[1:]

# 종류
print(s)


# 날씨
w = pd.read_csv('./weather_0020.csv')

w['지역(시)'] = w['지역(시)'].str.replace(pat='서울', repl='서울특별시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='광주', repl='광주광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='부산', repl='부산광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='대구', repl='대구광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='인천', repl='인천광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='대전', repl='대전광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='울산', repl='울산광역시',regex=False)
w['지역(시)'] = w['지역(시)'].str.replace(pat='제주특별자치도', repl='제주도',regex=False)


# 연도별로 쪼개기
years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012','2013','2014','2015','2016', '2017','2018','2019']

weather_list = []
prod_list = []
for i in years:
    prod_list.append(df[df['연도별']==i].reset_index(drop=True))
    weather_list.append(w[w['year']==int(i)].reset_index(drop=True))
prod_list[0]
```
