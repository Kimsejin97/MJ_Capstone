```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame as df
import numpy as np
import seaborn as sns
from matplotlib import rc
%matplotlib inline

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(data={'고추(양건)':[0.60,0.39,0.42,0.09],
             '고추(화건)':[0.64,0.36,0.42,0.09],
             '마늘':[0.13,-0.25,0.28,-0.14],
             '양파':[0.64,-0.66,0.67,-0.33]},
        index=['평년수입 + 연평균 가격', '평년수출 + 연평균 가격',
              '월별 수입 + 월별 가격', '월별 수출 + 월별 가격'])
```

```python
df
```

```python
plt.figure(figsize=(15,15))
plt.title("농산물 수입·수출량 가격과의 상관관계", y = 1.05, size = 20)
sns.heatmap(df, annot=True, fmt = '.2f', annot_kws = {'size' : 20},
            vmax=1, vmin=-1, linewidths=.1, cmap='RdBu')
plt.xticks(fontsize=18)
plt.yticks(fontsize=15, va='center')
plt.show()
```
