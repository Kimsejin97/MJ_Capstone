```python
weather1 = pd.read_excel('./날씨데이터/kma20200522174253492.xlsx')
weather2 = pd.read_excel('./날씨데이터/kma20200522174416436.xlsx')
weather3 = pd.read_excel('./날씨데이터/kma20200522174723598.xlsx')
weather4 = pd.read_excel('./날씨데이터/kma20200522175735661.xlsx')
weather5 = pd.read_excel('./날씨데이터/kma20200522175856809.xlsx')
weather6 = pd.read_excel('./날씨데이터/kma20200522180109240.xlsx')
```

```python
weather_data = weather[weather['지역(군)'].isin(['고창군', '고흥군', '의성군', '무안군', '해남군', '안동시'])]
weather_data['거래일자'] = pd.to_datetime(weather_data['거래일자'], format='%Y%m%d')
weather_data
#배추 - 전라북도 고창군
#무 - 전라북도 고창군
#마늘 - 전라남도 고흥군, 경상북도 의성군
#양파 - 전라남도 무안군
#건고추 - 전라남도 해남군, 경상북도 안동시
```

<h4>상관관계</h4>

<h4>8~10월 강수량은 무 생산량과 얼마만큼의 상관관계가 있는가?(1.1.1)</h4>

```python
data = weather_data[weather_data['지역(군)'] == '고창군']
data['거래일자'] = pd.to_datetime(data['거래일자'], format='%Y%m%d')
data = data[data['거래일자'].dt.month.isin(['08','09','10'])]
```

```python
data['년도'] = data['거래일자'].dt.year
data['월'] = data['거래일자'].dt.month
del data['거래일자']
monthly_weather = data.groupby(['년도','월'], as_index=False).mean()
monthly_weather
```

```python
monthly_weather = monthly_weather[['년도','월','강수량']]
monthly_weather
```

```python
output = radish[radish['지역']== '전라북도']
output
```

```python
data = pd.merge(monthly_weather, output)
```
