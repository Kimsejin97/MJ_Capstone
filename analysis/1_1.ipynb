```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h4>생산량 데이터</h4>

```python
cabbage = pd.read_csv('./데이터/KOSIS_재배면적_생산량/시도별/배추_생산량.csv', encoding='euc-kr') # 배추
radish = pd.read_csv('./데이터/KOSIS_재배면적_생산량/시도별/무_생산량.csv', encoding='euc-kr') # 무
pepper = pd.read_csv('./데이터/KOSIS_재배면적_생산량/시도별/건고추_생산량.csv', encoding='euc-kr') # 고추
garlic = pd.read_csv('./데이터/KOSIS_재배면적_생산량/시도별/마늘_생산량.csv', encoding='euc-kr') # 마늘
onion = pd.read_csv('./데이터/KOSIS_재배면적_생산량/시도별/양파_생산량.csv', encoding='euc-kr') # 양파
```

<h4>날씨 데이터</h4>

```python
weather = pd.read_csv('./데이터/weather.csv', encoding='euc-kr')
```

<h4>생산량 데이터 + 날씨 데이터</h4>

```python
cabbage_data = pd.merge(weather, cabbage, left_on=None)
cabbage_data
```

```python
radish_data = pd.merge(weather, radish, left_on=None)
radish_data
```

```python
pepper_data = pd.merge(weather, pepper, left_on=None)
pepper_data
```

```python
garlic_data = pd.merge(weather, garlic, left_on=None)
garlic_data
```

```python
onion_data = pd.merge(weather, onion, left_on=None)
onion_data
```

<h4>상관관계<h4>
