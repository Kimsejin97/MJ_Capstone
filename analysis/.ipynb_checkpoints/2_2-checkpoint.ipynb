<h2>2.2)농산물(고추,마늘,양파) 생산량이 가격에 영향을 끼치는가?</h2>

<h4>연도별 데이터 상관관계</h4>

```python
x = pepper_yang_year_data['생산량 (톤)']
y = pepper_yang_year_data['평균가격']

plt.plot(x, y, 'o')
plt.show()
```

```python
corr = pepper_yang_year_data[['생산량 (톤)', '평균가격']].corr(method='pearson')
corr#건고추(양건)생산량과 가격은 음의 상관관계
```

```python
x = pepper_hwa_year_data['생산량 (톤)']
y = pepper_hwa_year_data['평균가격']

plt.plot(x, y, 'o')
plt.show()
```

```python
corr = pepper_hwa_year_data[['생산량 (톤)', '평균가격']].corr(method='pearson')
corr#건고추(화건)생산량과 가격은 음의 상관관계
```

```python
x = garlic_year_data['생산량 (톤)']
y = garlic_year_data['평균가격']

plt.plot(x, y, 'o')
plt.show()
```

```python
corr = garlic_year_data[['생산량 (톤)', '평균가격']].corr(method='pearson')
corr#마늘 생산량과 가격은 음의 상관관계
```

```python
x = onion_year_data['생산량 (톤)']
y = onion_year_data['평균가격']

plt.plot(x, y, 'o')
plt.show()
```

```python
corr = onion_year_data[['생산량 (톤)', '평균가격']].corr(method='pearson')
corr
```

<h2>2.2.1) 고추의 전년산 생산량은 당해 연도 8월 가격에 영향을 끼치는가?</h2>

<h4>고추_양건</h4>

```python
output = pepper.iloc[:19,3] #2019년을 제외하고 연도별 공급량 추출
output
```

```python
price_yang = pepper_yang_price[pepper_yang_price['월']==8].reset_index() #8월만 추출
price_yang = price_yang.iloc[1:,3].reset_index(drop=True) #2000년을 제외하고 8월가격 추출
price_yang
```

```python
output = pd.DataFrame(output)
price_yang = pd.DataFrame(price_yang)
```

```python
data = pd.concat([output, price_yang], axis=1)
```

```python
corr = data.corr(method='pearson')
corr
```

<h4>고추_화건</h4>

```python
price_hwa = pepper_hwa_price[pepper_yang_price['월']==8].reset_index() #8월만 추출
price_hwa = price_hwa.iloc[1:,3].reset_index(drop=True) #2000년을 제외하고 8월가격 추출
price_hwa
```

```python
price_hwa = pd.DataFrame(price_hwa)
```

```python
data = pd.concat([output, price_hwa], axis=1)
```

```python
corr = data.corr(method='pearson')
corr
```
