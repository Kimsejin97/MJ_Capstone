```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rc
%matplotlib inline
```

```python
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
#년도, 평균가격
pepper_yang_year_price =  pd.read_csv('./데이터/KAMIS_가격/건고추_앙건_연도별_가격.csv')
pepper_hwa_year_price =  pd.read_csv('./데이터/KAMIS_가격/건고추_화건_연도별_가격.csv')
garlic_year_price = pd.read_csv('./데이터/KAMIS_가격/마늘_난지_연도별_가격.csv')
onion_year_price = pd.read_csv('./데이터/KAMIS_가격/양파_연도별_가격.csv')
```

```python
#양건
plt.figure(figsize=(20,15))
plt.plot(pepper_yang_year_price['년도'],pepper_yang_year_price['평균가격'], marker='o')
plt.xticks(np.arange(2000,2020, step=1),
           ["{}".format(i) for i in np.arange(2000,2020, step=1)])
plt.yticks(np.arange(5000,27000, step=1000),
           ["{}".format(i) for i in np.arange(5000,27000, step=1000)])
plt.title('고추_양건',fontsize=30, pad=20)
plt.xlabel('년도',fontsize=20)
plt.ylabel('평균가격',fontsize=20)
plt.show()
```

```python
#화건
plt.figure(figsize=(20,15))
plt.plot(pepper_hwa_year_price['년도'],pepper_hwa_year_price['평균가격'], marker='o')
plt.xticks(np.arange(2000,2020, step=1),
           ["{}".format(i) for i in np.arange(2000,2020, step=1)])
plt.yticks(np.arange(5000,27000, step=1000),
           ["{}".format(i) for i in np.arange(5000,27000, step=1000)])
plt.title('고추_화건',fontsize=30, pad=20)
plt.xlabel('년도',fontsize=20)
plt.ylabel('평균가격',fontsize=20)
plt.show()
```

```python
#양건화건 그래프 비교
plt.figure(figsize=(20,15))
plt.plot(pepper_yang_year_price['년도'],pepper_yang_year_price['평균가격'], marker='o',label='양건')
plt.plot(pepper_hwa_year_price['년도'],pepper_hwa_year_price['평균가격'], marker='o',label='화건')
plt.xticks(np.arange(2000,2020, step=1),
           ["{}".format(i) for i in np.arange(2000,2020, step=1)])
plt.yticks(np.arange(5000,27000, step=1000),
           ["{}".format(i) for i in np.arange(5000,27000, step=1000)])
plt.title('고추 화건,양건 가격추이',fontsize=30, pad=20)
plt.xlabel('년도',fontsize=20)
plt.ylabel('평균가격',fontsize=20)
plt.legend(fontsize=30,loc='lower right')
plt.show()
```

```python
#마늘
plt.figure(figsize=(20,15))
plt.plot(garlic_year_price['년도'],garlic_year_price['평균가격'], marker='o')
plt.xticks(np.arange(2000,2020, step=1),
           ["{}".format(i) for i in np.arange(2000,2020, step=1)])
plt.yticks(np.arange(500,7000, step=500),
           ["{}".format(i) for i in np.arange(500,8000, step=500)])
plt.title('마늘',fontsize=30, pad=20)
plt.xlabel('년도',fontsize=20)
plt.ylabel('평균가격',fontsize=20)
plt.show()
```

```python
#양파
plt.figure(figsize=(20,15))
plt.plot(onion_year_price['년도'],onion_year_price['평균가격'], marker='o')
plt.xticks(np.arange(2000,2020, step=1),
           ["{}".format(i) for i in np.arange(2000,2020, step=1)])
plt.yticks(np.arange(300,1500, step=100),
           ["{}".format(i) for i in np.arange(300,1500, step=100)])
plt.title('양파',fontsize=30, pad=20)
plt.xlabel('년도',fontsize=20)
plt.ylabel('평균가격',fontsize=20)
plt.show()
```
